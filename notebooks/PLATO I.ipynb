```python
from plato.environment import Environment
```

```python
config = {
  "whites":{
    "params":{
    "n_white_infantry":5,
    "n_white_vehicle":2,
    "n_white_drone":1},
    "init":"NW"
  },

  "blacks":{
    "params":{
    "n_black_infantry":5,
    "n_black_vehicle":2,
    "n_black_drone":1},
    "init":"SE"
  }
}
```

```python
sim = Environment(config, shape=(60,60))
```

```python
obs = sim.reset()
```

```python
for id_,w in sim.whites.items(): print('{:13}'.format(id_), w)
```

```python
for id_,w in sim.blacks.items(): print('{:13}'.format(id_), w)
```

```python
canvas = sim.render()
```
