```python
from plato.environment import Environment
```

```python
w = {"params":{
    "n_white_infantry":12,
    "n_white_vehicle":2,
    "n_white_drone":1}
    }

b = {"params":{
    "n_black_infantry":12,
    "n_black_vehicle":2,
    "n_black_drone":1}
    }
```

```python
from plato.objective.spatial.capture_objective import CaptureObjective
objectives = [CaptureObjective({'xy':[50,50],'radius':10})]
```

```python
sim = Environment({"whites":{**w, "init":"N"},"blacks":{**b, "init":"S"}}, shape=(100,100), objectives=objectives)
obs = sim.reset()
canvas = sim.render()
```
